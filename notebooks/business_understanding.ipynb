## Project Overview

This project focuses on analyzing historical rainfall data in India (1901–2015) to identify climate patterns and build predictive models for agricultural and urban planning applications.

The project follows the CRISP-DM (Cross-Industry Standard Process for Data Mining) methodology to ensure a structured and scalable data science workflow.

## Business Objectives

### Q1: Crop Planning (Business Problem)
Can we predict whether a future period (3 months ahead) will experience:
- A drought (rainfall < -20% of normal)
- Heavy rainfall (rainfall > +20% of normal)

This helps farmers optimize:
- Crop selection
- Irrigation planning
- Risk management

### Q2: Urban Planning (Research Problem)
Is there an increasing trend in extreme rainfall events over the last 30 years?

This helps determine whether:
- Current stormwater infrastructure is sufficient
- Cities need redesign for flood resilience

## Data Mining Goals

### For Q1:
- Build a classification model to predict:
  - Drought (Yes/No)
  - Heavy Rainfall (Yes/No)
- Prediction horizon: 3 months ahead

### For Q2:
- Identify extreme rainfall events (outliers)
- Analyze their frequency over time
- Detect trends using statistical methods

## Target Variable Definition

To answer Q1, new target variables will be created:

- Drought:
  Rainfall < 80% of historical average

- Heavy Rainfall:
  Rainfall > 120% of historical average

These thresholds represent a ±20% deviation from normal rainfall.

For Q2:
- Extreme events will be defined as statistical outliers
  (e.g., beyond 1.5 × IQR or top 5% rainfall values)

## Success Criteria

### Q1 (Prediction Model):
- Accuracy
- Precision & Recall (important for drought detection)
- F1-score

### Q2 (Trend Analysis):
- Statistically significant increase in extreme events
- Clear upward trend in frequency over time

## Assumptions and Constraints

- Historical rainfall patterns are indicative of future trends
- External climate factors (e.g., temperature, humidity) are not initially included
- Data quality may vary across years and regions
- The dataset may not fully capture micro-climate variations

## Risks

- Climate change may introduce non-stationarity in data
- Missing or inconsistent records
- Model overfitting due to temporal dependencies
- Lack of external features may reduce prediction accuracy
